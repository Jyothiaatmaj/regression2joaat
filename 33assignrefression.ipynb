```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/car_age_price.csv")
data
```

```python
corr_matrix =data.corr()
```

```python
sns.heatmap(corr_matrix, annot=True,cmap="YlGnBu")
plt.show()
```

```python
x = data['Year'].values.reshape(-1, 1)
y = data['Price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =0.2,random_state =42)
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
model =lin_regressor.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("MSE =",mean_squared_error(y_test,predictions))
print("R squared value = ",r2_score(y_test,predictions))
```

```python
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(predictions,color='red', label='Linear Regression Line')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Linear Regression: Year vs. Price')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso
import numpy as np
```

```python
data = pd.read_csv('/content/car_age_price.csv')
```

```python
X = data['Year'].values.reshape(-1, 1)
y = data['Price'].values
```

```python
lasso = Lasso()
lasso.fit(X, y)
```

```python
y_pred = lasso.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', linewidth=2, label='Lasso Regression')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Lasso Regression: Year vs. Price')
plt.legend()
plt.show()
```
